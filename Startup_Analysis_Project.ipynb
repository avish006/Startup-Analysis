```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/indian_startup_funding.csv')
```

```python
df
```

```python
df['Date']= df['Date'].fillna("2015-01-02")
```

```python
df['Investment Type']  = df['Investment Type'].fillna("Unknown")
df['Investment Type'].isna().sum()
```

```python
df['Investors']=df['Investors'].fillna("Unknown Investors")
df['Amount in USD']=df['Amount in USD'].fillna(df['Amount in USD'].median())
```

```python
df['Location']=df['Location'].replace("Bengaluru","Bangalore")
```

```python
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
```

```python
df['Industry'] = df['Industry'].fillna("Unknown Industry")
```

```python
df['Sub-vertical'] = df['Sub-vertical'].fillna(df['Sub-vertical'].mode()[0])
```

```python
df.isna().sum()
```

```python
# Clean up Industry names
df['Industry'] = df['Industry'].str.strip().str.lower()

# Map similar/duplicate values to consistent labels
industry_map = {
    'eCommerce': 'E-commerce',
    'eCommerce': 'Ecommerce',
    'eCommerce ': 'ECommerce',
    'eCommerce': 'ecommerce',
    'eCommerce': 'e-commerce',
    'eCommerce': 'e-Commerce',
    'eCommerce': 'E-Commerce',
    'fintech': 'fintech',
    'finance': 'fintech',
    'edu-tech': 'edtech',
    'edutech': 'edtech',
    'edu tech': 'edtech',
    'b2b': 'b2b',
    'technology': 'technology',
    'consumer internet': 'consumer internet',
    'transportation': 'transportation',
    'hospitality': 'hospitality',
    'health and wellness': 'health and wellness',
    'ecommerce':'e-commerce'
}

df['Industry'] = df['Industry'].replace(industry_map)
```

```python
# List of vague investor names to exclude
invalid_investors = [
    'unknown investors',
    'undisclosed investors',
    'undisclosed investor',
    'group of angel investors',
    'na',
    'n/a',
    '',
    None
]

# Clean text formatting just in case
df['Investors'] = df['Investors'].str.strip().str.lower()

# Remove invalid rows
df = df[~df['Investors'].isin(invalid_investors)]
```

```python
df.to_csv("cleaned_startup_data.csv")
```
